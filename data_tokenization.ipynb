```python
nlp = spacy.load("en_core_web_md")
pipe2 = []
for sentences in pipe1:
    tokens = []
    for sentence in sentences:
        doc = nlp(sentence)
        tokens.append([token.text for token in doc])
    pipe2.append(tokens)
```
