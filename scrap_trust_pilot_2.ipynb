```python
def get_rev_info(all_revs: list):
    rev_info = []
    i = 1
    for rev in all_revs:
        info = []
        URL = rev[1]
        # print(URL)
        info.append(rev[0])
        info.append(rev[1])
        page = requests.get(URL)
        # IP can be blocked
        sleep(2)

        soup = BeautifulSoup(page.content, "html.parser")
        # Get company Name
        results_name = soup.find("span", class_="typography_display-s__qOjh6")
        if results_name != None:
            info.append(results_name.text)
        else:
             info.append(np.nan)

        # Get the Overall score
        results1 = soup.find("p", class_="typography_body-l__KUYFJ")
        # print(results1.text)
        if results1 != None:
            info.append(results1.text)
        else: 
            info.append(np.nan)

        # Get the descrition 
        results2 = soup.find("div", class_="styles_container__9nZxD")
        # print(results2.text)
        if results2 != None:
            info.append(results2.text)
        else:
            info.append(np.nan)

        # Get the number of pages for each review
        results3 = soup.find("nav", class_="pagination_pagination___F1qS")

        if results3 != None:
            if len(results3.find_all("a", href=True)) > 3:
                max_page = results3.find_all("a", href=True)[-2].text
                # print(max_page)
                info.append(max_page)
            else:
                max_page = results3.find_all("a", href=True)[-1].text
                # print(max_page)
                info.append(max_page)
        else:
            info.append(np.nan)
        
        rev_info.append(info)
        i+=1
        # add checkpoint
        if i % 10 == 0:
            with open(f"rev/rev_info", "wb") as fp:
                pickle.dump(rev_info, fp)
            print("checkpoint :", i)

    return rev_info
```

```python
rev_1 = get_rev_info(all_revs[:1000])
```
