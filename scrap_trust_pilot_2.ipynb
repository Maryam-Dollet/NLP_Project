```python
def get_rev_info(all_revs: list, i: int = 0):
    # rev_info = []
    # i = 1
    if i != 0:
        with open(f"rev/rev_info_{i}", "rb") as fp:   # Unpickling
            rev_info = pickle.load(fp)
    else:
        rev_info = []
    for rev in all_revs:
        info = []
        URL = rev[1]
        # print(URL)
        info.append(rev[0])
        info.append(rev[1])
        page = requests.get(URL)
        # IP can be blocked
        sleep(2)

        soup = BeautifulSoup(page.content, "html.parser")
        # Get company Name
        results_name = soup.find("span", class_="typography_display-s__qOjh6")
        if results_name != None:
            info.append(results_name.text)
        else:
             info.append(np.nan)

        # Get the Overall score
        results1 = soup.find("p", class_="typography_body-l__KUYFJ")
        # print(results1.text)
        if results1 != None:
            info.append(results1.text)
        else: 
            info.append(np.nan)

        # Get the descrition 
        results2 = soup.find("div", class_="styles_container__9nZxD")
        # print(results2.text)
        if results2 != None:
            info.append(results2.text)
        else:
            info.append(np.nan)

        # Get the number of pages for each review
        results3 = soup.find("nav", class_="pagination_pagination___F1qS")

        if results3 != None:
            if len(results3.find_all("a", href=True)) > 2:
                max_page = results3.find_all("a", href=True)[-2].text
                # print(max_page)
                info.append(int(max_page))
            else:
                max_page = results3.find_all("a", href=True)[-1].text
                # print(max_page)
                info.append(int(max_page))
        else:
            info.append(np.nan)
        
        # get reviews on 2 pages
        results4 = soup.find_all("section", class_="styles_reviewContentwrapper__zH_9M")
        reviews = []
        if results4 != None:
            for res in results4:
                if res.find("div", class_="styles_reviewHeader__iU9Px") != None:
                    score = res.find("div", class_="styles_reviewHeader__iU9Px")["data-service-review-rating"]
                else:
                    score = np.nan
                a = res.find("a", class_="link_internal__7XN06")
                if a != None:
                    link = a["href"]
                    title = a.text
                else:
                    link = ""
                    title = np.nan
                if res.find('p', class_ = "typography_body-l__KUYFJ") != None:
                    txt = res.find('p', class_ = "typography_body-l__KUYFJ").text
                else:
                    txt = np.nan
                # print(link, title, txt)
                reviews.append(["https://fr.trustpilot.com" + link, score, title, txt])
            
            if type(max_page) != str:
                if int(max_page) > 1:
                    page = requests.get(URL+"?page=2")
                    soup2 = BeautifulSoup(page.content, "html.parser")
                    results5 = soup2.find_all("section", class_="styles_reviewContentwrapper__zH_9M")
                    for res in results5:
                        if res.find("div", class_="styles_reviewHeader__iU9Px") != None:
                            score = res.find("div", class_="styles_reviewHeader__iU9Px")["data-service-review-rating"]
                        else:
                            score = np.nan
                        a = res.find("a", class_="link_internal__7XN06")
                        if a != None:
                            link = a["href"]
                            title = a.text
                        else:
                            link = ""
                            title = np.nan
                        if res.find('p', class_ = "typography_body-l__KUYFJ") != None:
                            txt = res.find('p', class_ = "typography_body-l__KUYFJ").text
                        else:
                            txt = np.nan
                        # print(link, title, txt)
                        reviews.append(["https://fr.trustpilot.com" + link, score, title, txt])
        info.append(reviews)
        
        rev_info.append(info)
        i+=1
        # add checkpoint
        if i % 10 == 0:
            if os.path.isfile(f"rev/rev_info_{i-20}"):
                os.remove(f"rev/rev_info_{i-20}")
            with open(f"rev/rev_info_{i}", "wb") as fp:
                pickle.dump(rev_info, fp)
            print("checkpoint :", i)

    return rev_info
```
